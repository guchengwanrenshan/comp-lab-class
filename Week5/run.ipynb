```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import pandas as pd
```

```python
trr_file = "./adp_T300.trr"
gro_f = './adp.gro'
traj = md.load(trr_file,top= gro_f)
x,phi = md.compute_phi(traj, periodic=True, opt=True)
print(phi)
```

```python
y, psi = md.compute_psi(traj, periodic=True, opt=True)
print(psi)
```

```python
print(traj.time)
```

```python
plt.plot(traj.time, phi)
plt.xlabel('Time (ps)')
plt.ylabel('phi-angle')
plt.title('phi-angle&time')
plt.show()
```

```python
plt.plot(traj.time, psi)
plt.xlabel('Time (ps)')
plt.ylabel('psi-angle')
plt.title('psi-angle&time')
plt.show()
```

```python
H, xedges,yedges = np.histogram2d(phi.flatten(),psi.flatten(),bins=100,range=None, normed=False, weights=None)
H = -np.log(H)
H = H -np.min(H)
plt.imshow(H,interpolation='nearest',origin='lower',extent=[xedges[0],xedges[-1],yedges[0],yedges[-1]])
plt.xlabel('Psi(rad)')
plt.ylabel('Phi(rad)')
plt.colorbar()
plt.show()
```

```python
df_3= pd.read_csv('./T300/potential.xvg',skiprows=24,sep='\s+',header=None)
df_4= pd.read_csv('./T400/potential.xvg',skiprows=24,sep='\s+',header=None)
df_6= pd.read_csv('./T600/potential.xvg',skiprows=24,sep='\s+',header=None)
df_10= pd.read_csv('./T1000/potential.xvg',skiprows=24,sep='\s+',header=None)
```

```python
weight_300 = np.ones_like(df_3[1]) / len(df_3[1])
weight_400 = np.ones_like(df_4[1]) / len(df_4[1])
weight_600 = np.ones_like(df_6[1]) / len(df_6[1])
weight_1000 = np.ones_like(df_10[1]) / len(df_10[1])

plt.figure()

plt.hist(df_3[1], weights=weight_300, edgecolor='black', color='red', label='T=300 K')
plt.hist(df_4[1], weights=weight_400, edgecolor='black', color='yellow', label='T=400 K')
plt.hist(df_6[1], weights=weight_600, edgecolor='black', color='blue', label='T=600 K')
plt.hist(df_10[1], weights=weight_1000, edgecolor='black', color='green', label='T=1000 K')

plt.legend()
plt.xlabel("Potential (kJ/mol)", fontsize=10)
plt.ylabel("Frequency", fontsize=10)
#plt.savefig('replica_%s.png' dpi=500, bbox_inches='tight')
plt.show()
```
