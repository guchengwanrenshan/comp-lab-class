```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('./Data/1hz3_T310.run.25000000.energy.xvg',skiprows=29,sep='\s+',header=None)
file.columns = [ "t(ps)","Potential(kJ/mol)" ,"Kinetic En.(kJ/mol)", "Total Energy(kJ/mol)", "Temperature(K)","Pressure(bar)","Volume(nm^3)"]
file
```

```python
file.insert(1, 't(ns)', 0, allow_duplicates=False) # df.insert(loc, column, value, allow_duplicates=False)
ns=[]
for i in range(len(file)):
    file.iloc[i,1] = file.iloc[i,0]/1000
    if i % 10 == 0:
        ns.append(i)
print(ns)
df= file.loc[ns]
df
```

```python
x =  df['t(ns)']
y1 = df['Potential(kJ/mol)']
y2 = df['Kinetic En.(kJ/mol)']
y3 = df['Total Energy(kJ/mol)']
y4 = df['Temperature(K)']
y5 = df['Pressure(bar)']
y6 = df['Volume(nm^3)']
# figsize是画布尺寸，facecolor是背景颜色，num为图形编号或者名称
plt.set_title("Potential(kJ/mol),Kinetic En.(kJ/mol),Total Energy(kJ/mol),Temperature(K),Pressure(bar),Volume(nm^3)")
plt.figure(num=3,figsize=(16,24),facecolor='w')
plt.plot(x, y1)
plt.plot(x,y2,color='r',linewidth=3,linestyle='--')
plt.plot(x,y3,color='b',linewidth=3,linestyle='-.')
plt.plot(x,y4,color='y',linewidth=3,linestyle=':')
plt.plot(x,y5,color='m',linewidth=3,linestyle='solid')
plt.plot(x,y6,color='c',linewidth=3,linestyle='dashed')
plt.show()
```

```python
x_a =df['t(ns)']
y_a1 = df['Kinetic En.(kJ/mol)']
y_a2 = df['Temperature(K)']
fig = plt.figure(num=3,figsize=(16,24),facecolor='w')
ax1 = fig.add_subplot(111)
ax1.plot(x_a, y_a1,'g',linewidth=3,linestyle='solid')
ax1.set_ylabel('Kinetic En.(kJ/mol)')
ax1.set_title("Kinetic En.&Temperature")
ax2 = ax1.twinx()  # this is the important function
ax2.plot(x_a, y_a2, 'r',linewidth=3,linestyle='dashed')#supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
#ax2.set_xlim([0, np.e])
ax2.set_ylabel('Temperature(K)')
ax2.set_xlabel('t(ns)')
plt.show()
```
