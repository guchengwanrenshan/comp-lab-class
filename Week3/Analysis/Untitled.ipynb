```python
def get_pic(input):
    path = input
    f = open(path).read().splitlines()
    data =[]
    for i in f:
        if "s0 legend" in i :
            title = i.replace("s0 legend",'').replace('"','').replace('@','').replace(' ','')
            print(title)
        if "xaxis  label" in i :
            x = i.replace("xaxis  label",'').replace('"','').replace('@','').replace(' ','')
        if "yaxis  label" in i :
            y = i.replace("yaxis  label",'').replace('"','').replace('@','').replace(' ','')
        if i[0]!='@'and i[0]!='#':
            data.append(i.replace(' ',''))
    df = pd.DataFrame(data)
    df['time'] = df[0].map(lambda x:x.split('00000')[0])
    df['num'] = df[0].map(lambda x:x.split('00000')[1])
    
    fig = plt.figure(num=3,figsize=(16,24),facecolor='w')
    #fig.add_subplot(111)
    plt.figure(figsize=(20, 8), dpi=100)
    plt.plot(df['time'], df['num'], linewidth=1, color="orange", marker="o",label="density")
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(title, fontsize=20)
    plt.xlim(-1,101)
    #plt.ylim(900.000,1100.000)
    plt.xticks(np.linspace(0,100,11,endpoint=True))# ä¿®æ”¹çºµåæ ‡çš„åˆ»åº¦
    #yticks(np.linspace(900,1100,5,endpoint=True))
    # 2.4 å›¾åƒä¿å­˜
    plt.savefig("./"+input+".png")
    # 2.5 æ·»åŠ å›¾ä¾‹
    plt.legend(loc="best")
    # 3.å›¾åƒæ˜¾ç¤º
    plt.show()
    
```
