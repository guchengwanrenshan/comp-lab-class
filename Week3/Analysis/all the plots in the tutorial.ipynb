```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f1 = pd.read_csv('./density.xvg',skiprows=24,sep='\s+',header=None)
f1.columns = [ "Time (ps)","energy(kg/m^3)" ]
f1
```

```python
fig1 = plt.figure(figsize=(20, 10), dpi=100)              #创建空图
#x_label = [1,2,3,4,5,6]     #x轴的坐标
#y_label = [1,3,2,3,2,6]     #y轴坐标
plt.plot(f1['Time (ps)'],f1['energy(kg/m^3)'],color = 'r',linewidth=1.0,linestyle='--')  #构建折线图，可以设置线宽，颜色属性
plt.title("density")                 #设置标题，这里只能显示英文，中文显示乱码
plt.xlabel("time(ps)")            #设置y轴名称
plt.ylabel("energy(kg/m^3)")            #设置x轴名称
plt.show()   
plt.savefig("./density.png")
```

```python
f2 = pd.read_csv('./potential.xvg',skiprows=297,sep='\s+',header=None)
f2.columns = [ "Time (ps)","energy(kJ/mol)" ]
f2.drop(index = [688],inplace = True)
f2
```

```python
fig2 = plt.figure(figsize=(20, 10), dpi=100)              #创建空图
plt.xticks(np.linspace(0,900,19,endpoint=True))    #x轴的坐标
#y_label = [1,3,2,3,2,6]     #y轴坐标
plt.plot(f2['Time (ps)'],f2['energy(kJ/mol)'],color = 'g',linewidth=1.0,linestyle='-')  #构建折线图，可以设置线宽，颜色属性
plt.title("potential")                 #设置标题，这里只能显示英文，中文显示乱码
plt.xlabel("time(ps)")            #设置y轴名称
plt.ylabel("energy((kJ/mol)")            #设置x轴名称
plt.show() 
plt.savefig("./potential.png")
```

```python
f4=pd.read_csv('./rmsd.xvg',header=0,names=['Time(ns)','Ref:Equilibrated'],skiprows=18,sep='\s+')
f5=pd.read_csv('./rmsd_xtal.xvg',header=0,names=['Time(ns)','Ref:crystal'],skiprows=18,sep='\s+')
fig4 = plt.figure(figsize=(20, 10), dpi=100)       
plt.plot(f4['Time(ns)'],f4['Ref:Equilibrated'],label='Ref:Equilibrated')
plt.plot(f4['Time(ns)'],f5['Ref:crystal'],label='Ref:crystal')
plt.xlabel('Time (ns)')
plt.ylabel('RMSD(nm)')
plt.title('RMSD')
plt.legend()
plt.show() 
plt.savefig("./RMSD.png")
```

```python
f6=pd.read_csv('./gyrate.xvg',header=0,names=['Time(ps)','Rg','Rg\sX','Rg\sY','Rg\sZ'],skiprows=27,sep='\s+')
f6
```

```python
fig5 = plt.figure(figsize=(20, 10), dpi=100)       
plt.plot(f6['Time(ps)'],f6['Rg'],label='Rg')
plt.plot(f6['Time(ps)'],f6['Rg\sX'],label='Rg\sX')
plt.plot(f6['Time(ps)'],f6['Rg\sY'],label='Rg\sY')
plt.plot(f6['Time(ps)'],f6['Rg\sZ'],label='Rg\sZ')
plt.xlabel('Time(10ps)')
plt.ylabel('Rg(nm)')
plt.title('Radius of gyration')
plt.legend()
plt.show()
plt.savefig("./Rg.png")
```

```python
f7=pd.read_csv('./temperature.xvg',header=0,names=['Time(ps)','Temperature(K)'],skiprows=24,sep='\s+')
fig6 = plt.figure(figsize=(20, 10), dpi=100) 
plt.plot(f7['Time(ps)'],f7['Temperature(K)'],label='Temperature(K)',color = 'b',linewidth=1.0,linestyle='-') #构建折线图，可以设置线宽，颜色属性
plt.title("Temperature")                 #设置标题，这里只能显示英文，中文显示乱码
plt.xlabel("time(ps)") #设置x轴名称
plt.ylabel('Temperature(K)') #设置y轴名称
plt.legend(loc="best")
plt.show() 
plt.savefig("./temperature.png")
```
